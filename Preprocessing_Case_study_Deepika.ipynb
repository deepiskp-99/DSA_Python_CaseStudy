**2.Rows: Check for duplicate rows in the dataset, if any, and remove them.
Columns: Identify and drop duplicate columns, if any, based on their values.**

```python
# Remove duplicate rows
duplicates = data.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
data = data.drop_duplicates()

# Check for duplicate columns
duplicate_columns = data.columns[data.T.duplicated()]
print(f"Duplicate columns: {duplicate_columns.tolist()}")
data = data.loc[:, ~data.T.duplicated()]
```

**3.Identify missing values in each column.
Handle missing values:
For numerical columns, use imputation techniques like mean/median imputation.
For categorical columns, fill with mode.
Document your approach for each feature with missing data.**

```python
data.isnull().sum()
```

```python
for col in data.columns:
    if data[col].isnull().sum() > 0:
        if data[col].dtype in ['int64', 'float64']:   # Numerical
            # Use median (better for skewed data)
            data[col].fillna(data[col].median(), inplace=True)
        else:  # Categorical
            # Fill with mode (most frequent value)
            data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

**4.Identify all numerical columns (excluding the target variable SalePrice).
Scale these features using techniques like Min-Max scaling or Standard scaling.**

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Step 1: Identify numerical columns (excluding SalePrice)
num_cols = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
if "Sale Price" in num_cols:
    num_cols.remove("Sale Price")

print("Numerical columns to scale:", num_cols)

# Step 2a: Standard Scaling (z-score normalization)
standard_scaler = StandardScaler()
data[num_cols] = standard_scaler.fit_transform(data[num_cols])

# Alternative: Step 2b: Min-Max Scaling (values between 0 and 1)
minmax_scaler = MinMaxScaler()
data[num_cols] = minmax_scaler.fit_transform(data[num_cols])


data.head()
```

**5.Identify all categorical columns in the dataset.
Apply appropriate encoding techniques:
Use One-Hot Encoding for nominal categories.
Use Label Encoding for ordinal categories, if applicable.**


```python
#Encoding Categorical Variables
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

# One-Hot Encoding for nominal features
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
```

**6.Perform an outlier detection analysis on numerical variables (e.g., using the IQR method).
Remove outliers from these features if they are not representative of typical house prices.**

```python
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = data.shape[0]
    data = data[(data[col] >= lower) & (data[col] <= upper)]
    after = data.shape[0]
    print(f"Removed {before - after} outliers from column '{col}'")

print(f"\nDataset shape after outlier removal: {data.shape}")
```
